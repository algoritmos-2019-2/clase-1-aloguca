#  Funciones

* ¿Por qué son ventajosas las funciones para tener en sus programas?

Porque asi el código puede reducirse de una manera muy favorable ya que solo defines esa funcion una vez y cuando quieras volver a realizar las mismas acciones, en lugar de escribir nuevamente todo el codigo, lo unico que haces es llamar a la funcion.
* ¿Cuándo se ejecuta el código en una función: cuando se define la función o cuando se llama a la función?

Cuando se llama a la funcion, porque al llamarla, el compilador ejecuta el codigo definido en la funcion.
* ¿Qué enunciado crea una función?

def funcion ()
* ¿Cuál es la diferencia entre una función y una llamada de función?

Una funcion es la parte del codigo donde esta definida, donde se crea la funcion y llamada de funcion es cuando nombras la funcion para que ejecute el codigo dentro de ella.
* ¿Cuántos ámbitos globales hay en un programa de Python? ¿Cuántos ámbitos locales?

Habran tantos como uno quiera, no hay limite.
* ¿Qué sucede con las variables en un ámbito local cuando se devuelve la llamada a la función?

Son ejecutadas ya que al hacer la llamada a la funcion, indicas que se ejecute el codigo de la funcion donde se encuentan las variables locales.
* ¿Qué es un valor de retorno? ¿Puede un valor de retorno ser parte de una expresión?

Es un comando que tiene la funcion de retornar a alguna parte del codigo, depende de lo que se le indque retornar. Si puede ser parte de una expresion.
* Si una función no tiene una declaración de retorno, ¿cuál es el valor de retorno de una llamada a esa función?

Nada, si no tiene indicado nada entonces return no te retornara a nada y por lo tanto no afectara al codigo.
* ¿Cómo puede forzar una variable en una función para referirse a la variable global?

Indicando a la variable que retorne la variable global.
* ¿Cuál es el tipo de datos de None?

Son los que significan nada o ninguno, lo que hacen es indicar eso mismo, sin embargo no debe confundirse con que el valor sea cero.
* ¿Qué hace la declaración import random?

Importa una funcion de otra libreria.
* Si tuviera una función llamada randint() en un módulo llamado random, ¿cómo lo llamaría después de importar random?

Para mandarla a llamar se deberia escribir de la siguiente manera: nombre_del_modulo.nombre_de_la_funcion ()

#  Listas y Tuplas

* ¿Qué es []?

Son parentesis cuadrados donde se define el contenido de una lista.
* ¿Cómo asignaría el valor 'hola' como el tercer valor en una lista almacenada? en una variable llamada spam? (Supongamos que el spam contiene [2, 4, 6, 8, 10].) 

Pueden ser de dos maneras dependiendo lo que se quiera, si se quiere cambiar el tercer valor de la lista, se haria de la siguiente manera:
spam [2] = 'hola'
En el caso de que quiera añadir un valor en el tercer valor de la lista, seria de la siguiente forma:
spam.insert (2, "hola")

Para las siguientes tres preguntas, digamos que spam contiene la lista [2, 4, 6, 8, 10].
* ¿Qué evalúa spam[int ('3' * 2) / 11]?

Marcara un error ya que los indices de la lista deben ser enteros o segmentos y esta operacion nos da un flotante.
* ¿Qué evalúa spam[-1]?

Nos dara el valor de 10, ya que el compilador comienza a contar a partir de 0, entonces el -1 lo toma como el ultimo indice  de la lista.
* ¿Qué evalúa el spam[: 2]?

Nos dara los valores [2, 4] ya que estas indicando que muestre los indices desde el inicio hasta el indice 2, que daria igual si se evaluara spam [0:2]

Para las siguientes tres preguntas, digamos que bacon contiene la lista [3.14, 'cat', 11, 'cat', True].

* ¿Qué evalúa bacon.index('cat')?

Nos indica el indice en la lista donde se encuentra cat, en este caso como esta dos veces y como el compilador lee de arriba hacia abajo, te devolvera el primer lugar donde lo haya encontrado.
* ¿Qué evalúa bacon.append(99)?

Añade el valor de 99 a la lista, cabe destacar que el valor se añadira al final de la lista.
* ¿Qué hace bacon.remove('cat')?

Lo que hara es quitar de la lista cat, pero de igual forma, quitara unicamente el primer cat que encuentre.
* ¿Cuáles son los operadores para la concatenación de listas y la replicación de listas?

Se utiliza el operador de la suma.
* ¿Cuál es la diferencia entre los métodos de lista append() e insert()?

append () lo que hace es añadir solo un valor a la lista, mientras que insert () lo que hace es añadir un valor a la lista pero en el indice donde se indico ser añadido.
* ¿Cuál es la diferencia entre listas y tuplas?

La diferencia mas destacable de las tuplas es que son inmutables.
